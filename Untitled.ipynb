```python
import pandas as pd

def read_college_data(df):
    """Read college data from a DataFrame."""
    colleges = []
    for _, row in df.iterrows():
        colleges.append({
            'College ID': int(row['College ID']),
            'College Name': row['College Name'],
            'Branch': row['Branch'],
            'Minimum Marks': int(row['Minimum Marks'])
        })
    return colleges

def get_eligible_colleges(score, colleges):
    """Get a list of eligible colleges based on the score."""
    eligible_colleges = [college for college in colleges if score >= college['Minimum Marks']]
    
    if eligible_colleges:
        print("You are eligible for the following colleges:")
        for college in eligible_colleges:
            print(f"College ID: {college['College ID']}, Name: {college['College Name']}, Branch: {college['Branch']}")
    else:
        print("No, you have not scored enough marks to qualify for any college.")

def main():
    """Main function to execute the program."""
    # Load the dataset
    df = pd.read_csv('csv_preprocessing_2.csv')
    
    # Read the college data from the DataFrame
    colleges = read_college_data(df)
    
    # Input score
    user_score = int(input("Enter your score: "))
    
    # Get eligible colleges based on user score
    get_eligible_colleges(user_score, colleges)

if __name__ == "__main__":
    main()
```

```python
# first 
import pandas as pd

def read_college_data(df):
    """Read college data from a DataFrame."""
    colleges = []
    for _, row in df.iterrows():
        colleges.append({
            'College ID': row['College ID'],  # Do not convert to int if it contains non-integer values
            'College Name': row['College Name'],
            'Branch': row['Branch'],
            'Minimum Marks': pd.to_numeric(row['Minimum Marks'], errors='coerce')  # Convert to numeric, handle errors
        })
    return colleges

def get_eligible_colleges(score, colleges):
    """Get a list of eligible colleges based on the score."""
    eligible_colleges = [college for college in colleges if college['Minimum Marks'] is not None and score >= college['Minimum Marks']]
    
    if eligible_colleges:
        print("You are eligible for the following colleges:")
        for college in eligible_colleges:
            print(f"College ID: {college['College ID']}, Name: {college['College Name']}, Branch: {college['Branch']}")
    else:
        print("No, you have not scored enough marks to qualify for any college.")

def main():
    """Main function to execute the program."""
    # Load the dataset
    df = pd.read_csv('csv_preprocessing_2.csv')
    
    # Read the college data from the DataFrame
    colleges = read_college_data(df)
    
    # Input score
    user_score = int(input("Enter your score: "))
    
    # Get eligible colleges based on user score
    get_eligible_colleges(user_score, colleges)

if __name__ == "__main__":
    main()
```
